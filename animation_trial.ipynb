```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Generate data for the line
x = np.linspace(0, 10, 100)  # X-axis values
y = np.sin(x)                # Y-axis values (for example, a sine wave)

# Set up the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 10)           # Set x-axis range
ax.set_ylim(-1, 1)           # Set y-axis range
line, = ax.plot([], [], lw=2)  # Line object to be animated

# Initialization function: Clear the line
def init():
    line.set_data([], [])
    return line,

# Update function: Draw the line incrementally
def update(frame):
    line.set_data(x[:frame], y[:frame])
    return line,

# Create the animation
ani = FuncAnimation(
    fig, update, frames=len(x), init_func=init, blit=True, interval=20
)

# Display the animation
plt.show()
```
